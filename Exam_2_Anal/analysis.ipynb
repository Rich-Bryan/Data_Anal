```python
import pandas as pd
```

```python
filename = "exam_2.csv"
df = pd.read_csv(filename)
```

```python
# the statistics 
print(df.describe())
```

```python
import matplotlib.pyplot as plt

# Plot
plt.figure(figsize=(10, 6))
plt.hist(df, bins=30, edgecolor="black", alpha=0.7)
plt.title("Grade Distribution", fontsize=16)
plt.xlabel("Grades", fontsize=14)
plt.ylabel("Students", fontsize=14)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Extracting descriptive statistics for annotation
mean_grade = df["grade"].mean()
median_grade = df["grade"].median()
min_grade = df["grade"].min()
max_grade = df["grade"].max()
```

```python
# Creating a histogram with annotations for key statistics
plt.figure(figsize=(10, 6))
plt.hist(df, bins=30, color="skyblue", edgecolor="black", label="Grade Distribution")

# Adding lines and annotations for statistics
plt.axvline(mean_grade, color="red", linestyle="--", linewidth=1.5, label=f"Mean: {mean_grade:.2f}")
plt.axvline(median_grade, color="green", linestyle="--", linewidth=1.5, label=f"Median: {median_grade:.2f}")
plt.axvline(min_grade, color="purple", linestyle="--", linewidth=1.5, label=f"Min: {min_grade}")
plt.axvline(max_grade, color="orange", linestyle="--", linewidth=1.5, label=f"Max: {max_grade}")

# Customizing the plot
plt.title("Grade Distribution with Key Statistics", fontsize=16)
plt.xlabel("Grades", fontsize=14)
plt.xticks(range(0, 110, 10)) 
plt.ylabel("Frequency", fontsize=14)
plt.legend(fontsize=12)
plt.grid(axis="y", linestyle="--")
plt.tight_layout()
plt.show()
```
